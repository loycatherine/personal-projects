```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Input, Dense, Flatten, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import load_img, img_to_array

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")

from pathlib import Path
from IPython.display import display
from PIL import Image
import pickle
```

```python
train_data_folder = Path("data/train")
train_df = pd.read_csv("data/train.csv")

train_df["filepath"] = ""
for index, row in train_df.iterrows():
    row["filepath"] = train_data_folder / row["filename"]
```

```python
train_df.head()
```

```python
count_row = train_df.shape[0]
count_row
```

```python
# convert all the images into arrays
train_images_arr = []
for index, row in train_df.iterrows():
    z = image.load_img(row["filepath"], target_size=(300,300))
    r = image.img_to_array(z)
    r = preprocess_input(r)
    train_images_arr.append(r)
    
train_images_arr = np.array(train_images_arr)

train_labels = []
for index, row in train_df.iterrows():
    if row["class"] == "with_mask":
        train_labels.append(1.0)
    else:
        train_labels.append(0.0)
        
train_labels = np.array(train_labels, dtype="float32")
```

```python
train_labels
```

```python
test_data_folder = Path("data/test")
test_df = pd.read_csv("data/test.csv")

test_df["filepath"] = ""
for index, row in test_df.iterrows():
    row["filepath"] = test_data_folder / row["filename"]
```

```python
# convert all the images into arrays
test_images_arr = []
for index, row in test_df.iterrows():
    z = image.load_img(row["filepath"], target_size=(300,300))
    r = image.img_to_array(z)
    r = preprocess_input(r)
    test_images_arr.append(r)
    
test_images_arr = np.array(test_images_arr)

test_labels = []
for index, row in test_df.iterrows():
    if row["class"] == "with_mask":
        test_labels.append(1.0)
    else:
        test_labels.append(0.0)
        
test_labels = np.array(test_labels, dtype="float32")
```

```python
count_row_test = test_df.shape[0]
count_row_test
```

```python
# Sampling and visually inspecting 20 images:

import random
for i in random.sample(range(1178), 5):
    display(Image.open(train_df['filepath'][i]))
```

```python
# Checking for class imbalance in dataset

num_mask= 0
num_no_mask = 0

labels = ["Without Mask", "With Mask"]

for row in range(1178):
    if train_df["class"][row] == 'with_mask':
        num_mask +=1
    else:
        num_no_mask +=1

for row in range(194):
    if test_df["class"][row] == 'with_mask':
        num_mask +=1
    else:
        num_no_mask +=1

num_classes = [num_no_mask, num_mask]
     
plt.bar(labels, num_classes)
plt.title("No. of images per class in our dataset")
plt.xlabel("Class")
plt.ylabel("No. of images")
plt.show()
```

```python
generator = ImageDataGenerator(
    rotation_range=20,
    zoom_range=0.15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    horizontal_flip=True,
    fill_mode="nearest")
```

```python
i = Input(shape=(300, 300, 3))

pretrained_model = MobileNetV2(weights="imagenet", include_top=False, input_tensor = i, pooling="average")

for layer in pretrained_model.layers:
    layer.trainable = False

x = pretrained_model.output
x = GlobalAveragePooling2D()(x)
x = Flatten()(x)
x = Dense(1024, activation="relu")(x)
x = Dropout(0.2)(x)
x = Dense(1, activation="sigmoid")(x)
```

```python
model = Model(pretrained_model.input, x)
model.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
batch_size = 32
steps_per_epoch = train_images_arr.shape[0] // batch_size
validation_steps = test_images_arr.shape[0] // batch_size
```

```python
r = model.fit(
    generator.flow(train_images_arr, train_labels, batch_size=batch_size),
    steps_per_epoch = steps_per_epoch,
    validation_data = (test_images_arr, test_labels),
    validation_steps = validation_steps,
    epochs = 20
)
```

```python
with open("./model.pickle",'wb') as f:
    pickle.dump(r.history, f)
```

```python
model.save('model')
```

```python
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```

```python
plt.plot(r.history['accuracy'], label='acc')
plt.plot(r.history['val_accuracy'], label='val_acc')
plt.legend()
```

```python
score, acc = model.evaluate(test_images_arr, test_labels, batch_size=32)
print("Test loss:", score)
print("Test accuracy:", acc)
```

```python
import seaborn as sn

from sklearn.metrics import confusion_matrix, classification_report
```

```python
def predict(model, x, y, batch_size, model_name):
  print("Predicting with", model_name)
  probabilities = model.predict(x, batch_size=batch_size)
  predictions = probabilities.round(decimals=0).astype(int)
  
  report = classification_report(y, predictions, target_names=["0", "1"], digits=3)
  print(report)

  cm = confusion_matrix(y, predictions)
  df_cm = pd.DataFrame(cm, index=["0", "1"], columns=["0", "1"])
  sn.heatmap(df_cm, cmap=plt.cm.Blues, annot=True, fmt="d")
  plt.xlabel("Predicted")
  plt.ylabel("Actual")
  plt.title("Confusion matrix for {}".format(model_name))
  plt.show()
```

```python
predict(model, test_images_arr, test_labels, 32, "MobileNetV2")
```
